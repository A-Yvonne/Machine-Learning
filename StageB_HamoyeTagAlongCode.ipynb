Linear model for T2 (x) and T6 (y)

```python
x = energy[['T2']]
y = energy[['T6']]
```

Split into training and test dataset

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE = {round(rmse, 3)}")
```

MinMaxScaler Normalisation and Multiple Linear Regression

```python
energyapp = energy.drop(['date', 'lights'], axis = 1)
```

```python
scaler = MinMaxScaler()

normalised_energyapp = pd.DataFrame(scaler.fit_transform(energyapp), columns = energyapp.columns)

energy_app = normalised_energyapp.drop(columns = ['Appliances'], axis=1)
energy_target = normalised_energyapp['Appliances']
```

```python
x_train, x_test, y_train, y_test = train_test_split(energy_app, energy_target, test_size = 0.3, random_state=42)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

Mean Absolute Error in 3dp

```python
mse = mean_absolute_error(y_test, y_pred)
print(f"MSE = {round(mse, 3)}")
```

Root Mean Squared Error in 3dp

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
print(f"RMSE = {round(rmse, 3)}")
```
